```python
def jaccard_similarity(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    return float(intersection) / union
```

```python
import pandas as pd
```

```python
db_example = pd.read_pickle("./db_example_100.pkl")
```

```python
db_example
```

```python
db_example['ocr_dict_lookup_list'] = db_example.apply(lambda x: [item for sublist in x['ocr_dict_lookup'] for item in sublist], axis=1)
```

```python
import numpy as np
def ocr_percentage(myrow, colname="ocr_dict_lookup_list", min_num_letters=4):
    myrow_array = np.array(myrow['ocr_dict_lookup_list']).astype(np.int)
    ocr_perc = len(myrow_array[myrow_array >= min_num_letters]) / len(myrow_array[abs(myrow_array) >= min_num_letters])
    return ocr_perc*100
```

```python
db_example['ocr_dict_perc'] = db_example.apply(ocr_percentage, args=["ocr_dict_lookup_list", 4], axis=1)

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.plot(db_example['str_similarity'], db_example['ocr_dict_perc'], 'ko')
plt.grid()
plt.xticks(size=24)
plt.yticks(size=24)
plt.xlabel("String similarity measure", size=32)
plt.ylabel("Words found in the dictionary (%)", size=32)
plt.xlim(0.0, 1.0)
plt.ylim(0, 100)
plt.show()
```
