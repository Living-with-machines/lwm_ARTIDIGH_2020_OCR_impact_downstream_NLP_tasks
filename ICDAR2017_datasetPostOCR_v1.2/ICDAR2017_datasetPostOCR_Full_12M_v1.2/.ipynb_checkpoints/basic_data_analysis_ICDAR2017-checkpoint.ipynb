# Basic Data Analysis of ICDAR2017 

```python
from glob import glob
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
list_monographs = glob("./eng_monograph/*.txt")
list_periodical = glob("./eng_periodical/*.txt")
```

```python
list_monographs.sort()
list_periodical.sort()
```

```python
all_monographs = []
for lmono in list_monographs:
    mono_fio = open(lmono, "r")
    one_mono = mono_fio.readlines()
    all_monographs.append([os.path.basename(lmono), 
                           one_mono[0].split("[OCR_toInput] ")[1], 
                           one_mono[1].split("[OCR_aligned] ")[1], 
                           one_mono[2].split("[ GS_aligned] ")[1]])
all_monographs = pd.DataFrame(all_monographs, columns=["ID", "OCR_toInput", "OCR_aligned", "GS_aligned"])
all_monographs.head()
```

```python
all_periodical = []
for lper in list_periodical:
    per_fio = open(lper, "r")
    one_perio = per_fio.readlines()
    all_periodical.append([os.path.basename(lper), 
                           one_perio[0].split("[OCR_toInput] ")[1], 
                           one_perio[1].split("[OCR_aligned] ")[1], 
                           one_perio[2].split("[ GS_aligned] ")[1]])
all_periodical = pd.DataFrame(all_periodical, columns=["ID", "OCR_toInput", "OCR_aligned", "GS_aligned"])
all_periodical.head()
```

```python
num_hashes_monographs = all_monographs["GS_aligned"].apply(lambda x: x.count("#")/float(len(x))*100.).values
num_hashes_periodical = all_periodical["GS_aligned"].apply(lambda x: x.count("#")/float(len(x))*100.).values
```

```python
comparable_lims = False

plt.figure(figsize=(20, 10))

plt.subplot(2, 1, 1)
plt.title("Monographs", size=30)
plt.plot(num_hashes_monographs, "k")
plt.ylabel("% of OCR error", size=20)
plt.xticks(size=20)
plt.yticks(size=20)
plt.grid()

if comparable_lims:
    plt.ylim(0, 40)
    plt.xlim(0, 750)

plt.subplot(2, 1, 2)
plt.title("Periodicals", size=30)
plt.plot(num_hashes_periodical, "k")
plt.ylabel("% of OCR error", size=20)
plt.xlabel("#documents", size=20)
plt.xticks(size=20)
plt.yticks(size=20)
plt.grid()
#plt.ylim(0, 40)
#plt.xlim(0, 750)

plt.tight_layout()
plt.show()
```
